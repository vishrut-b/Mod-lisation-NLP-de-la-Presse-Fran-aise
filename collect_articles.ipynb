````python
import requests
import json

API_KEY = "2b9fd40f5d204b0e8f6db4da23f2a5c1"
url = "https://newsapi.org/v2/everything"

params = {
    "q": "actualité",        # French word to get news
    "language": "fr",        # French language filter
    "pageSize": 100,         # Max articles per page
    "page": 1,
    "apiKey": API_KEY
}

all_articles = []

while True:
    response = requests.get(url, params=params)
    data = response.json()

    if data.get("status") != "ok":
        print("Error from NewsAPI:", data)
        break

    articles = data.get("articles", [])
    if not articles:
        break

    all_articles.extend(articles)
    print(f"Fetched {len(articles)} articles, total so far: {len(all_articles)}")

    if len(articles) < params["pageSize"]:
        # Last page reached
        break

    params["page"] += 1

# Save articles to a JSON file
with open("french_news_articles.json", "w", encoding="utf-8") as f:
    json.dump(all_articles, f, ensure_ascii=False, indent=2)

print(f"Total articles saved: {len(all_articles)}")
````

````python
import re

def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)
    # Remove HTML tags if any
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters except accents
    text = re.sub(r'[^\w\sàâçéèêëîïôûùüÿñæœ]', '', text)
    # Normalize spaces
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

cleaned_texts = [clean_text(t) for t in texts]
````

#### Here, the problem is that using ```newsapi```gives me access a 100 articles, but truncated. This is not very helpful as some articles can take time and space to build context, rendering the process of extracting meaning a difficult task.

#### A better option I have found on the internet is to utilise the ```newspaper``` module in Python.

````python
from newspaper import Article
import time

def clean_text(text):
    Basic cleaning to remove newlines etc.
    return text.replace('\n', ' ').strip() if text else ""

# Load NewsAPI articles JSON
with open("french_news_articles.json", encoding="utf-8") as f:
    articles = json.load(f)

enriched_articles = []

for idx, article in enumerate(articles):
    url = article.get("url")
    print(f"Processing article {idx+1}/{len(articles)}: {url}")

    full_text = ""
    if url:
        try:
            news_article = Article(url, language='fr') # Download raw HTML from the URL
            news_article.download()                    # Extract and clean article text + metadata from HTML
            news_article.parse()                       # Now this contains just the article’s main text
            full_text = clean_text(news_article.text)  # An optional cleaning step.
        except Exception as e:
            print(f"Failed to scrape full article: {e}")

    # Fallback to API snippet if scraping failed or no text extracted
    if not full_text:
        title = article.get("title") or ""
        description = article.get("description") or ""
        content = article.get("content") or ""
        full_text = clean_text(f"{title}. {description}. {content}")

    # Add the full text to the article dictionary
    article["full_text"] = full_text
    enriched_articles.append(article)

    # Avoid hammering servers too fast
    time.sleep(1)  # adjust if you want slower/faster scraping

# Save enriched articles with full_text field
with open("enriched_french_news_articles.json", "w", encoding="utf-8") as f:
    json.dump(enriched_articles, f, ensure_ascii=False, indent=2)

print(f"Done! Enriched {len(enriched_articles)} articles with full text where available.")
````

````python
import json
import re
import time
from newspaper import Article
import nltk
from langdetect import detect
from nltk.tokenize import sent_tokenize

# Download once:
# nltk.download('punkt')

# 1) Define global advert‐boilerplate markers & regexes
AD_MARKERS = [
    "Publicité", "Découvrir l’offre", "Mise à jour", "Modèle concerné",
    "offre chez", "Désactivez votre bloqueur", "Source :", 
    "Pour ne manquer aucune actualité", "Vous n’avez pas eu le temps"
]

# Regex to remove common boilerplate blocks in one go
BOILERPLATE_PATTERNS = [
    re.compile(
        r"Vous n’avez pas eu le temps de suivre l’actualité hier.*?newsletter Frandroid",
        flags=re.IGNORECASE | re.DOTALL
    ),
    re.compile(
        r"Mise à jour\s*\|.*?L'offre n'est plus disponible.*?(?=\.)",
        flags=re.IGNORECASE
    ),
    # add more if you see other repeated chunks
]

def clean_text(text: str) -> str:
    """Basic whitespace cleanup + strip boilerplate blocks + remove ad sentences."""
    if not text:
        return ""

    # 1) Flatten whitespace
    txt = text.replace('\r', ' ').replace('\n', ' ').strip()

    # 2) Remove known boilerplate blocks via regex
    for pat in BOILERPLATE_PATTERNS:
        txt = pat.sub("", txt)

    # 3) Split into sentences and drop any that contain markers
    sentences = sent_tokenize(txt, language='french')
    good = [
        s for s in sentences
        if not any(marker in s for marker in AD_MARKERS)
    ]

    # 4) Rejoin and trim
    cleaned = " ".join(good).strip()
    return cleaned

def is_valid_french(text: str, min_length: int = 200) -> bool:
    """Quick check: enough length and detected language is French."""
    if len(text) < min_length:
        return False
    try:
        return detect(text) == 'fr'
    except:
        return False

# Load your raw API articles
with open("french_news_articles.json", encoding="utf-8") as f:
    api_articles = json.load(f)

enriched = []
seen_texts = set()

for idx, art in enumerate(api_articles, 1):
    url = art.get("url")
    print(f"[{idx}/{len(api_articles)}] Fetching {url}")

    full = ""
    # 1) Try newspaper scraping
    if url:
        try:
            na = Article(url, language='fr')
            na.download(); na.parse()
            full = clean_text(na.text)
        except Exception as e:
            print("  scrape failed:", e)

    # 2) Fallback to API snippet
    if not full:
        snippet = " ".join(filter(None, (
            art.get("title"), art.get("description"), art.get("content")
        )))
        full = clean_text(snippet)

    # 3) Validation: length, language, duplicates
    if not is_valid_french(full):
        print("  dropped: too short or non-French")
        continue
    if full in seen_texts:
        print("  dropped: duplicate text")
        continue

    # 4) Accept and record
    seen_texts.add(full)
    art["full_text"] = full
    enriched.append(art)

    time.sleep(1)   # polite delay

# Save only the cleaned, valid articles
with open("enriched_french_news_articles.json", "w", encoding="utf-8") as f:
    json.dump(enriched, f, ensure_ascii=False, indent=2)

print(f"\nDone: {len(enriched)} articles retained out of {len(api_articles)}")
````

### Day 2 of data collection:
Since my ```newsapi``` plan allows only 100 articles to be collected per day, I will run this again to further add 100 more articles into the same data file. That would increase our sample size as well as capture trends that might run across several days. 

````python
import json, requests, time
from datetime import datetime, timedelta

# 1) Load your already‐enriched articles & URLs
with open("enriched_french_news_articles.json", encoding="utf-8") as f:
    enriched = json.load(f)
seen_urls = {art["url"] for art in enriched if art.get("url")}

# 2) Set up NewsAPI params—with a date window
API_KEY = "2b9fd40f5d204b0e8f6db4da23f2a5c1"
endpoint = "https://newsapi.org/v2/everything"

# We'll page through a moving 1-day window until we get ~100 new articles
new_articles = []
window_end = datetime.utcnow()      # start from right now
days_back = 1                       # 1-day windows
max_needed = 100                    # stop once we have ~100 new

while len(new_articles) < max_needed:
    window_start = window_end - timedelta(days=days_back)
    params = {
        "q": "actualité",
        "language": "fr",
        "from": window_start.isoformat(timespec='seconds') + "Z",
        "to":   window_end.isoformat(timespec='seconds') + "Z",
        "pageSize": 100,
        "page": 1,
        "sortBy": "publishedAt",
        "apiKey": API_KEY,
    }

    resp = requests.get(endpoint, params=params).json()
    hits = resp.get("articles", [])
    if not hits:
        # nothing in that window → go back another day
        window_end = window_start
        continue

    # filter out URLs we’ve already processed
    fresh = [a for a in hits if a.get("url") not in seen_urls]
    for art in fresh:
        seen_urls.add(art["url"])
        new_articles.append(art)
        if len(new_articles) >= max_needed:
            break

    print(f"Window {window_start.date()} → {window_end.date()}: fetched {len(hits)} hits, {len(fresh)} new")
    window_end = window_start           # shift window backwards
    time.sleep(1)                       # be polite to the API

# 3) At this point `new_articles` holds up to 100 new, unique items.
print(f"Total new fetched: {len(new_articles)}")

# 4) Now run your exact same enrich+clean loop on new_articles,
#    and append the results to enriched, then overwrite your JSON file.
#    (Reuse your newspaper/clean_text/is_valid_french logic.)

# Example stub:
# for art in new_articles:
#     …scrape via newspaper…
#     …clean_text & validation…
#     enriched.append(art)

# Finally save:
# with open("enriched_french_news_articles.json", "w", encoding="utf-8") as f:
#     json.dump(enriched, f, ensure_ascii=False, indent=2)
````

````python
from datetime import datetime, timedelta
from nltk.tokenize import sent_tokenize

# From your previous cell
# These are some of the unwanted phrases detected upon visual inspection.
# They were repetetive, and mostly advertisements.
AD_MARKERS = [
    "Publicité", "Découvrir l’offre", "Mise à jour", "Modèle concerné",
    "offre chez", "Désactivez votre bloqueur", "Source :",
    "Pour ne manquer aucune actualité", "Vous n’avez pas eu le temps"
]

# The phrases below, similar to the ones above, were found to be repetetive and unnecessary.
BOILERPLATE_PATTERNS = [
    re.compile(
        r"Vous n’avez pas eu le temps de suivre l’actualité hier.*?newsletter Frandroid",
        flags=re.IGNORECASE | re.DOTALL
    ),
    re.compile(
        r"Mise à jour\s*\|.*?L'offre n'est plus disponible.*?(?=\.)",
        flags=re.IGNORECASE
    ),
    # I might add more patterns as I discover them.
]


def clean_text_advanced(text: str) -> str:
    if not text:
        return ""
    txt = text.replace('\r', ' ').replace('\n', ' ').strip()
    for pat in BOILERPLATE_PATTERNS:
        txt = pat.sub("", txt)
    sents = sent_tokenize(txt, language='french')
    return " ".join(s for s in sents if not any(m in s for m in AD_MARKERS)).strip()

def is_valid_french(text: str, min_length: int = 200) -> bool:
    if len(text) < min_length:
        return False
    try:
        return detect(text) == 'fr'
    except:
        return False

# Load already-enriched articles and track URLs to avoid duplicates.
with open("enriched_french_news_articles.json", "r", encoding="utf-8") as f:
    enriched = json.load(f)
seen_urls = { art["url"] for art in enriched if art.get("url") }

# Fetch new article URLs via NewsAPI (this time using different keywords for greater sample size)
API_KEY = "2b9fd40f5d204b0e8f6db4da23f2a5c1"
endpoint = "https://newsapi.org/v2/everything"
keywords = ["actualité", "politique", "économie", "culture", "technologie"]

new_meta = []
for kw in keywords:
    params = {
        "q":            kw,
        "language":     "fr",
        "pageSize":     100,
        "page":         1,
        "sortBy":       "publishedAt",
        "apiKey":       API_KEY
    }
    resp = requests.get(endpoint, params=params).json()
    hits = resp.get("articles", [])
    fresh = [a for a in hits if a.get("url") not in seen_urls]
    for art in fresh:
        new_meta.append(art)
        seen_urls.add(art["url"])
    print(f"● '{kw}': {len(hits)} hits, {len(fresh)} new URLs")
    time.sleep(1)  # respect rate limits

print(f"\nTotal unique new URLs to process: {len(new_meta)}")

# Scrape & clean each new URL, validate, and append to `enriched`.
added = []
for idx, art in enumerate(new_meta, 1):
    url = art.get("url")
    print(f"[{idx}/{len(new_meta)}] Scraping {url}")
    full = ""
    try:
        na = Article(url, language='fr')
        na.download(); na.parse()
        full = clean_text_advanced(na.text)
    except Exception as e:
        print("  scrape failed:", e)

    if not full:  # fallback to API snippet if scraping gave nothing
        snippet = " ".join(filter(None, (art.get("title"), art.get("description"), art.get("content"))))
        full = clean_text_advanced(snippet)

    if not is_valid_french(full):
        print("  dropped: too short or not French")
        continue

    art["full_text"] = full
    enriched.append(art)
    added.append(art)
    time.sleep(1)

# Save appended articles to the same previous JSON file from Day 1
with open("enriched_french_news_articles.json", "w", encoding="utf-8") as f:
    json.dump(enriched, f, ensure_ascii=False, indent=2)

print(f"\nDone! Added {len(added)} new articles; total now {len(enriched)}.")
````
